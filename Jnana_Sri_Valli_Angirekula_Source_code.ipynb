```python
df.drop(columns=['license', 'house_rules'], inplace=True)
```

```python
df['reviews per month'].fillna(0, inplace=True)
df['price'].fillna('0', inplace=True)
df['service fee'].fillna('0', inplace=True)


df['price'] = df['price'].astype(str).str.replace(r'[$,]', '', regex=True)
df['service fee'] = df['service fee'].astype(str).str.replace(r'[$,]', '', regex=True)
```

```python
df['price'] = pd.to_numeric(df['price'])
df['service fee'] = pd.to_numeric(df['service fee'])

# 4. Replace placeholder 0s with the median of the non-zero values
price_median = df['price'][df['price'] != 0].median()
service_fee_median = df['service fee'][df['service fee'] != 0].median()

df['price'] = df['price'].replace(0, price_median)
df['service fee'] = df['service fee'].replace(0, service_fee_median)

# --- Step 5: Convert 'last_review' to datetime ---
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
numerical_cols_to_impute = [
    'minimum nights', 'number of reviews', 'review rate number',
    'calculated host listings count', 'availability 365', 'Construction year',
    'lat', 'long'
]
for col in numerical_cols_to_impute:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Convert construction_year to int64 (since it's a year)
df['Construction year'] = df['Construction year'].astype('int64')

# Impute remaining small object column NaNs with mode imputation
object_cols_to_impute = [
    'NAME', 'host_identity_verified', 'host name', 'neighbourhood group',
    'neighbourhood', 'country', 'country code', 'instant_bookable',
    'cancellation_policy'
]
for col in object_cols_to_impute:
    # Check if the mode is available before trying to impute
    if not df[col].mode().empty:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)

# --- Final Inspection ---
print("--- Final DataFrame Info After Preprocessing ---")
df.info()

print("\n--- Final Missing Value Check ---")
print(df.isnull().sum().to_markdown())
```

```python
import matplotlib.pyplot as plt

# Use the neighbourhood_listings Series calculated in cell b32f34cc
neighbourhood_listings = df['neighbourhood group'].value_counts()

plt.figure(figsize=(10, 6))
neighbourhood_listings.plot(kind='bar')
plt.title('Number of Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
